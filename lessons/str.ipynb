Индексация - процесс обращения к конкретному объету внутри итерируемова объекта. Любое обращение к индексам работает через `*объект*[*индекс*]`

Итерируемый объект - объект который хранит в себе последовательность объетов (т.е. <u>строка</u> - последовательность букв заключенных в ковычки)

```python
"Hello world"
# Индексация выглядит следующим образом
#0123456789 10
print("Hello world"[10])
print("Hello world"[0])
# print("Hello world"[11]) # ошибка
```

Индексовый срез - когда мы задаем начальный индекс и конечный индекс через `:`. И чтение идет от начального индекса <u>включительно</u> до конечного <u>исключительно</u>

```python
print("Hello world"[0:5])
print("Hello world"[0:6])
print("Hello world"[0:7])
```

Обратное индексирование - обратный процесс индексации, когда мы идем не с лева-направо, справа-налево. Соответвенно, индекс начинает тоже обратный отсчёт от 0, т.е. -1, -2, ... , -N *длина объекта*



```python
print("STRING"[-6])
print("STRING"[-6:-3])
print("STRING"[-3:-6]) # Ничего не получим тк шаг идет всегда вправо
```
