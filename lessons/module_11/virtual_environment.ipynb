# Виртуальное окружение

**Виртуальное окружение** — это инструмент, который позволяет создавать изолированные среды для проектов на Python. Каждое окружение имеет свои собственные зависимости, библиотеки и версии пакетов, что помогает избежать конфликтов между ними. Это особенно полезно, когда вы работаете над несколькими проектами одновременно, каждый из которых может требовать разные версии одних и тех же библиотек.

```python
!python -m venv venv_name # если папки проекта ещё не существует
!python -m venv . # если папка существует и мы хотим активировать виртуальное окружение в уже существующей папке
!source Scripts/activate  # активация виртуального окружения на Windows
```

# venv
Стандартная бибилотека которая позволяет создавать виртуальное окружение

# poetry
Бибилотека которая позволяет создавать виртуальное окружение, при этом не теряя зависимостей как например venv, а также можно управлять пакетами через команды poetry install, poetry update и т.д. Также установка происходит по конечным хэш-суммам изначальной установки библиотек, что позволяет установить библиотеки ровно таким образом как они были установлены у хозяина гит-ветки

```python
!pip install poetry(pip3.12 install poetry)
!poetry init # Нажимаем Enter - до тех пор пока виртуальное окружение не будет создано(python3.12 -m poetry init)
!poetry shell # Команда для входа в созданное виртуальное окружение(python3.12 run)
```
