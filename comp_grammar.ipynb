# Компьютер|Ноутбук

По содержимому Компьютер == Ноутбук

CPU (Основной процессор): Занимается вычислением в целом(Математики, Физики(в играх), Логикой(if-else) и пр.)
GPU*(Видеокарта): Занимается обработкой и работой с пиксилями экрана.
Если видеокарта в ноутбуке/компьютере отсутствует, то скорее всего процессор имеет встроенное видео-ядро

Видеокарта в ноутбуках это урезанная версия карт для компьютера
